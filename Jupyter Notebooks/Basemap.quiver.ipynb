```python
import numpy as np
import matplotlib.pyplot as plt
import netCDF4 as nc
from mpl_toolkits.basemap import Basemap
```

```python
PressureLevelsDS = nc.Dataset("/Users/macbook/PycharmProjects/20121015/108/20121015_00_ecmwf_ensemble_forecast.PRESSURE_LEVELS.EUR_LL10.108.pl.nc")
EQPT_DS = nc.Dataset("/Users/macbook/PycharmProjects/20121015/108/20121015_00_ecmwf_ensemble_forecast.EQPT_derived.EUR_LL10.108.ml.nc")
```

```python
U_velocity_isobaric1=np.array(PressureLevelsDS.variables['U_velocity_isobaric'][0,20,10,:,:])
V_velocity_isobaric1=np.array(PressureLevelsDS.variables['V_velocity_isobaric'][0,20,10,:,:])
```

```python
EQPT = np.array(EQPT_DS.variables['Equivalent_potential_temperature_hybrid'][0,20,10,:,:])
```

```python
longitude=np.array(PressureLevelsDS['lon'][:])
latitude=np.array(PressureLevelsDS['lat'][:])
(lon, lat) = np.meshgrid(longitude, latitude)
```

```python
fig, ax = plt.subplots(figsize=(30,12))
map = Basemap(projection='merc',llcrnrlat= latitude[-1],urcrnrlat= latitude[0],resolution='l',  llcrnrlon=longitude[0],urcrnrlon=longitude[-1])
map.quiver(lon,lat,U_velocity_isobaric1,V_velocity_isobaric1,latlon=True)
map.drawcoastlines(linewidth=0.1)
map.drawmapboundary(fill_color='aqua')
map.fillcontinents(color='coral',lake_color='aqua')
map.drawcountries(color='k',linewidth=0.5)
map.drawparallels(np.arange(30,90,5),labels=[False,True,True,False])
map.drawmeridians(np.arange(-60,40,10),labels=[True,False,False,True])
plt.show()
```
